# Nuevas Features


1. **Ruta**: Combina origen y destino
2. **Categoria Duracion**: Clasifica los viajes entre Corto (<1.5h), Medio (1.5h-3h) y Largo (>3h)
4. **Índice Severidad**: Qué tan grave es el retraso comparado con la duración del viaje (Retraso salida / Duración viaje). Un valor alto indica que el retraso afecta mucho al viaje.
5. **Recuperación**: Diferencia entre retraso de salida y llegada. Positivo = Recuperó tiempo / Negativo = Se retrasó más
6. **Tasa Cancelación (%)**: Para comparar estaciones grandes con pequeñas.

```python
# 1. Ruta
df['Ruta'] = df['Departure station'] + " -> " + df['Arrival station']

# 2. Categoria Duracion
etiquetas = ['Corto (<1.5h)', 'Medio (1.5h-3h)', 'Largo (>3h)']
df['Categoria Duracion'] = pd.cut(df['Average travel time (min)'],
                                  bins=[0, 90, 180, 1000],
                                  labels=etiquetas)

# 3. Índice de Severidad
df['Indice Severidad'] = df['Average delay of late departing trains (min)'] / df['Average travel time (min)']

# 4. Recuperación
df['Recuperacion'] = df['Average delay of late departing trains (min)'] - df['Average delay of late arriving trains (min)']

# 5. Tasa de Cancelación
df['Tasa Cancelacion'] = (df['Number of cancelled trains'] / df['Number of expected circulations']) * 100
```

# Preguntas de Negocio



1. ¿Cuáles son las rutas más problemáticas para los viajeros debido a su mayor cantidad de cancelaciones?

El análisis identifica la conexión Bordeaux St Jean ↔ Paris Montparnasse como el eje más crítico de la red, ocupando los dos primeros lugares del ranking y rozando las 2,000 cancelaciones en el trayecto de ida.

Se observa un claro patrón de fallo estructural bidireccional: las rutas problemáticas no son aisladas, sino que afectan a la línea completa (ida y vuelta), como se evidencia también en los corredores de Rennes y Marseille. Adicionalmente, la estación Paris Montparnasse se revela como un 'cuello de botella' sistémico, estando involucrada en la mitad del Top 10 (5 de las 10 rutas), lo que sugiere que los problemas operativos en este hub irradian cancelaciones hacia múltiples destinos.

```python
top_cancelaciones = df.groupby('Ruta')['Number of cancelled trains'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_cancelaciones.values, y=top_cancelaciones.index, palette="Reds_r")
plt.title('Top 10 Rutas con Mayor Número de Trenes Cancelados')
plt.xlabel('Total de Trenes Cancelados')
plt.ylabel('Ruta')
plt.show()
```

No existe correlación, ya que el coeficiente es cercano a 0. Además en el gráfico se observa que no importa si hay 10 trenes o 800 trenes programados (eje X), el retraso promedio (eje Y) se mantiene casi igual (cercano a cero).

3. ¿Son los viajes de larga distancia más propensos a sufrir retrasos graves que los trayectos cortos?

El análisis confirma que la duración del viaje es un factor crítico de riesgo. Los viajes de larga distancia (> 3h) no solo presentan un retraso mediano mayor, sino que sufren de una alta volatilidad operativa.

Mientras que los viajes cortos (< 1.5h) muestran un comportamiento estable y predecible, los viajes largos acumulan incidencias, resultando en una dispersión de datos mucho mayor y una propensión significativa a sufrir retrasos extremos (superiores a 100 minutos).

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Categoria Duracion', y='Average delay of late arriving trains (min)', palette="Set2")
plt.title('Distribución de Retrasos según la Duración del Viaje')
plt.ylabel('Minutos de Retraso (Llegada)')
plt.show()
```

4. ¿En qué epocas del año se producen mayores retrasos?

El análisis de series temporales revela una fuerte estacionalidad en la eficiencia operativa. Se identificó una Temporada Alta de incidencias claramente marcada en el periodo estival (Junio-Julio), donde los retrasos promedio se disparan hasta un 50% más que la media anual, probablemente impulsados por el flujo turístico y factores climáticos.

```python
estacionalidad = df.groupby(df['Period'].dt.month)['Average delay of all arriving trains (min)'].mean()

plt.figure(figsize=(10, 5))
sns.lineplot(data=estacionalidad, marker='o', color='green', linewidth=2.5)

plt.title('Estacionalidad de los Retrasos (Promedio Histórico por Mes)')
plt.xlabel('Mes')
plt.ylabel('Retraso Promedio (min)')

plt.xticks(ticks=range(1, 13), labels=[calendar.month_abbr[i] for i in range(1, 13)])

plt.grid(True)
plt.show()
```

5. ¿Han aumentado los retrasos por causas ajenas a la empresa (como clima o huelgas) en los últimos años?

Al analizar la tendencia anual de causas externas, no se observa un empeoramiento progresivo del servicio. Entre 2015 y 2019, el impacto se mantuvo estable en un rango alto, lo que sugiere que la infraestructura tiene una vulnerabilidad constante a factores ambientales y sociales.

Sin embargo, el año 2020 es un dato atípico (outlier) que presenta una caída histórica del 63% en los retrasos acumulados. Esta anomalía no refleja una mejora operativa interna, sino que es consecuencia directa de la reducción drástica del tráfico ferroviario debido a la pandemia de COVID-19, lo que disminuyó la exposición de los trenes a incidencias externas.

```python
tendencia_externa = df.groupby(df['Period'].dt.year)['Delay due to external causes'].sum().reset_index()

tendencia_externa.rename(columns={'Period': 'Year'}, inplace=True)

plt.figure(figsize=(10, 6))
grafico = sns.barplot(data=tendencia_externa, x='Year', y='Delay due to external causes', palette="Blues")

plt.title('Tendencia Anual: Impacto de Causas Externas (Clima, Huelgas, etc.)', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Minutos Totales de Retraso Acumulado', fontsize=12)

for p in grafico.patches:
    grafico.annotate(f'{int(p.get_height()):,}',
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'center',
                     xytext = (0, 9),
                     textcoords = 'offset points')

plt.show()
```

6. ¿La mayoría de los retrasos son culpa de la gestión interna o de factores externos?

Al realizar una auditoría de responsabilidad sobre los retrasos, se observa que los factores internos y de gestión operativa representan la gran mayoría de las incidencias, acumulando un 73.3% del total.

Si bien las Causas Externas (26.7%) figuran como la categoría individual más impactante, la suma de fallos en Infraestructura (25.5%), Gestión de Tráfico (19.2%) y Material Rodante (18.1%) evidencia que la principal oportunidad de mejora reside dentro de la propia organización, especialmente en el mantenimiento de vías y trenes, y no depende exclusivamente de factores del entorno.

```python
causas = df[[
    'Delay due to external causes',
    'Delay due to railway infrastructure',
    'Delay due to traffic management',
    'Delay due to rolling stock',
    'Delay due to station management and reuse of material',
    'Delay due to travellers taken into account'
]].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 8))
desfase = (0.1, 0, 0, 0, 0, 0)
plt.pie(causas, labels=causas.index, autopct='%1.1f%%', startangle=140, explode=desfase, shadow=True)
plt.title('Distribución de las Causas de Retraso (Total Acumulado)')
plt.show()
```

7. ¿En qué estaciones los trenes suelen salir ya con retrasos importantes respecto a la duración del viaje?

El análisis de severidad revela que Reims es la estación crítica número uno, presentando el mayor índice de desproporción entre el retraso sufrido y la duración programada del viaje (> 0.27). Esto significa que, porcentualmente, los retrasos en Reims impactan mucho más drásticamente la experiencia del pasajero que en otras estaciones.

Es alarmante observar a Paris Nord en segunda posición; al ser uno de los principales hubs ferroviarios de Europa, su alta severidad indica un problema sistémico de congestión que afecta no solo a rutas locales, sino a la eficiencia de la red troncal. Estas estaciones requieren una auditoría operativa prioritaria para identificar cuellos de botella en la fase de salida.

```python
top_severidad = df.groupby('Departure station')['Indice Severidad'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_severidad.values, y=top_severidad.index, palette="viridis")
plt.title('Top 10 Estaciones con Mayor Severidad de Retraso (Relativo a la duración)')
plt.xlabel('Índice de Severidad Promedio')
plt.show()
```

8. ¿Qué destinos actúan como "cuello de botella" acumulando la mayor cantidad de trenes retrasados?

El análisis identifica una saturación crítica centralizada en la capital. Las estaciones de Paris Lyon y Paris Montparnasse actúan como los grandes 'embudos' del sistema ferroviario, superando cada una los 50,000 trenes retrasados acumulados.

Lo más alarmante es la brecha abismal entre estos dos líderes y el resto: las estaciones parisinas acumulan más del triple de incidencias que el tercer lugar (Lyon Part Dieu). Esto sugiere que la red francesa sufre de una dependencia excesiva de sus hubs en París; si estas estaciones colapsan, bloquean la fluidez de toda la red nacional.

```python
bottleneck = df.groupby('Arrival station')['Number of trains late on arrival'].sum().sort_values(ascending=False).head(15)

plt.figure(figsize=(12, 7))
sns.barplot(x=bottleneck.values, y=bottleneck.index, palette="magma")
plt.title('Top 15 Estaciones de Destino con más Trenes Retrasados (Cuellos de Botella)')
plt.xlabel('Cantidad Total de Trenes Retrasados al Llegar')
plt.show()
```

9. ¿Son capaces los trenes de recuperar tiempo durante el viaje si salen con retraso?

Los datos demuestran que la red ferroviaria carece de elasticidad para absorber incidencias: los trenes NO son capaces de recuperar tiempo durante el viaje.

La distribución del histograma está claramente desplazada hacia la izquierda (valores negativos), con un pico de frecuencia que indica que lo más común es perder entre 10 y 30 minutos adicionales durante el trayecto si ya se salió tarde. Esto valida la existencia de un 'Efecto Bola de Nieve': un retraso inicial en la salida no se corrige, sino que tiende a agravarse sistemáticamente antes de llegar al destino, evidenciando una rigidez operativa que impide acelerar o ajustar la marcha para compensar demoras.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Recuperacion'], bins=30, kde=True, color='purple')
plt.axvline(0, color='red', linestyle='--') # Línea de referencia en 0
plt.title('Distribución de la Recuperación de Tiempo en Ruta')
plt.xlabel('Minutos Recuperados (Positivo = Recuperó tiempo / Negativo = Perdió más tiempo)')
plt.show()
```

10. ¿Se ha vuelto el servicio más inestable (con más cancelaciones) en alguna estación específica a lo largo de los años?

El análisis evolutivo revela una pérdida de fiabilidad sistémica en los últimos años. Mientras que el periodo 2015-2017 se caracterizó por una operación robusta y estable (tasas de cancelación cercanas al 0-2% en todas las estaciones), se identifican dos puntos de quiebre estructurales idénticos en todos los hubs:

- El pico de 2018: Un aumento generalizado (~8%) asociado a inestabilidad laboral (huelgas).
- El colapso de 2020: Un máximo histórico donde la tasa de cancelación se disparó hasta el 17.5% en Paris Est.

Esto demuestra que la red no tiene problemas aislados por estación, sino que es altamente vulnerable a shocks externos nacionales, reaccionando todas las estaciones principales al unísono ante crisis sanitarias o sociales.

```python
top_estaciones = df.groupby('Departure station')['Number of expected circulations'].sum().nlargest(4).index

df_top = df[df['Departure station'].isin(top_estaciones)].copy()

df_top['Year'] = df_top['Period'].dt.year

g = sns.relplot(
    data=df_top,
    x="Year",
    y="Tasa Cancelacion",
    col="Departure station",
    kind="line",
    col_wrap=2,
    height=3,
    aspect=1.5,
    marker='o',
    errorbar=None
)

g.fig.suptitle('Evolución de la Tasa de Cancelación en las Principales Estaciones', y=1.02)
g.set_axis_labels("Año", "Tasa de Cancelación (%)")

for ax in g.axes.flat:
    ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))

plt.show()
```
